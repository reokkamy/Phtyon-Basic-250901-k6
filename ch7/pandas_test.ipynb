```python
# í•™ìƒë“¤ì˜ ì´ë¦„, í•™ë…„, ì ìˆ˜ ë°ì´í„°
import pandas as pd
student_data = {
    'ì´ë¦„': ['ê¹€íŒë‹¤', 'ì´íš¨ë¦¬', 'ìœ ì¬ì„', 'ê°•í˜¸ë™'],
    'í•™ë…„': [2, 4, 1, 3],
    'ì ìˆ˜': [85, 95, 70, 88]
}

# ë”•ì…”ë„ˆë¦¬ë¥¼ ì‚¬ìš©í•˜ì—¬ DataFrame ìƒì„±
df = pd.DataFrame(student_data)

# ìƒì„±ëœ DataFrame ì¶œë ¥
print(df)
```

```python
#  ë°ì´í„° ì„ íƒí•˜ê¸° (Selecting Data)

# DataFrameì—ì„œ ì›í•˜ëŠ” ì—´ì´ë‚˜ í–‰ì„ ì„ íƒí•˜ëŠ” ê²ƒì€ ê°€ì¥ ê¸°ë³¸ì ì¸ ì‘ì—…ì…ë‹ˆë‹¤.
# 'ì´ë¦„' ì—´ë§Œ ì„ íƒí•˜ê¸° (Series í˜•íƒœë¡œ ë°˜í™˜ë¨)
names = df['ì´ë¦„']
print("--- 'ì´ë¦„' ì—´ ì„ íƒ ---")
print(names)

# 'ì´ë¦„'ê³¼ 'ì ìˆ˜' ë‘ ê°œì˜ ì—´ ì„ íƒí•˜ê¸° (DataFrame í˜•íƒœë¡œ ë°˜í™˜ë¨)
# [[...]] ì²˜ëŸ¼ ëŒ€ê´„í˜¸ë¥¼ ë‘ ë²ˆ ì‚¬ìš©í•©ë‹ˆë‹¤.
name_scores = df[['ì´ë¦„', 'ì ìˆ˜']]
print("\n--- 'ì´ë¦„', 'ì ìˆ˜' ì—´ ì„ íƒ ---")
print(name_scores)
```
