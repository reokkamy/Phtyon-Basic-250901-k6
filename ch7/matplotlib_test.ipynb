```python
# Í∞ÑÎã®Ìïú ÏÇ¨Ïö© ÏòàÏãú: Í∫æÏùÄÏÑ† Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
# Í∞ÄÏû• Í∏∞Î≥∏Ï†ÅÏù∏ Í∫æÏùÄÏÑ† Í∑∏ÎûòÌîÑ(Line Plot)Î•º Í∑∏Î¶¨Îäî Í≥ºÏ†ïÏùÄ Îã§ÏùåÍ≥º Í∞ôÏäµÎãàÎã§.
#
# 1. Îç∞Ïù¥ÌÑ∞ Ï§ÄÎπÑ: xÏ∂ïÍ≥º yÏ∂ïÏóê Ìï¥ÎãπÌïòÎäî Îç∞Ïù¥ÌÑ∞Î•º Ï§ÄÎπÑÌï©ÎãàÎã§.
# (Ï£ºÎ°ú Î¶¨Ïä§Ìä∏ÎÇò NumPy Î∞∞Ïó¥ ÏÇ¨Ïö©)
# 2. ÌîåÎ°Ø ÏÉùÏÑ±: plt.plot() Ìï®ÏàòÎ•º ÏÇ¨Ïö©Ìï¥ Îç∞Ïù¥ÌÑ∞Î•º Í∑∏ÎûòÌîÑÎ°ú Í∑∏Î¶ΩÎãàÎã§.
# 3. Í∑∏ÎûòÌîÑ Íæ∏ÎØ∏Í∏∞ (ÏÑ†ÌÉù ÏÇ¨Ìï≠):
# plt.title(), plt.xlabel(), plt.ylabel() Îì±ÏúºÎ°ú Ï†úÎ™©Í≥º Ï∂ï Ïù¥Î¶ÑÏùÑ Ï∂îÍ∞ÄÌï©ÎãàÎã§.
# 4. Í∑∏ÎûòÌîÑ Î≥¥Ïó¨Ï£ºÍ∏∞: plt.show() Ìï®ÏàòÎ•º Ìò∏Ï∂úÌïòÏó¨ ÌôîÎ©¥Ïóê Í∑∏ÎûòÌîÑÎ•º ÌëúÏãúÌï©ÎãàÎã§.

import matplotlib.pyplot as plt

# 1. Îç∞Ïù¥ÌÑ∞ Ï§ÄÎπÑ (xÏ∂ï: Ïõî, yÏ∂ï: ÏõîÎ≥Ñ Îß§Ï∂ú)
months = [1, 2, 3, 4, 5, 6]
sales = [150, 220, 280, 250, 310, 350]

# 2. Í∫æÏùÄÏÑ† Í∑∏ÎûòÌîÑ ÏÉùÏÑ±
plt.plot(months, sales)

# 3. Í∑∏ÎûòÌîÑ Íæ∏ÎØ∏Í∏∞
plt.title("Monthly Sales Trend") # Ï†úÎ™© Ï∂îÍ∞Ä
plt.xlabel("Month")              # xÏ∂ï Ïù¥Î¶Ñ Ï∂îÍ∞Ä
plt.ylabel("Sales (in millions)") # yÏ∂ï Ïù¥Î¶Ñ Ï∂îÍ∞Ä

# 4. Í∑∏ÎûòÌîÑÎ•º ÌôîÎ©¥Ïóê ÌëúÏãú
plt.show()
```
